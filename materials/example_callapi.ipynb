## Call Wisconet Wrappers

### Python - example

```python
import requests

url = "https://connect.doit.wisc.edu/ag_forecasting_api/ag_models_wrappers/wisconet"
params = {
    "forecasting_date": "2025-07-15",
    "risk_days": 7
}
headers = {"accept": "application/json"}

response = requests.get(url, params=params, headers=headers)
print(response.json())
```
